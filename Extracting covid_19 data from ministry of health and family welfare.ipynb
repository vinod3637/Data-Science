```python
import urllib.request as url
import re
import warnings
warnings.filterwarnings('ignore')
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np

import random
random.seed(123)
```

```python
url="https://www.mohfw.gov.in/"
wc=requests.get(url).content
soup=bs(wc,"html.parser")
print(soup)

```

```python
web = {'name_of_state/ut':[],'total_conformed_cases':[], 'Dischared':[],"Death":[]}
for i in soup.findAll("tr")[1:33]:
    #web["s.no"].append(i.findAll("td")[0].text)
    web['name_of_state/ut'].append(i.findAll("td")[1].text)
    web['total_conformed_cases'].append(i.findAll("td")[2].text)
    web['Dischared'].append(i.findAll("td")[3].text)
    web["Death"].append(i.findAll("td")[4].text)
```
