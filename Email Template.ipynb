# Task -2

Check and fetch email address from the email body section if any present and store it in another CSV. 

```python
a=email["Email Boby"]
```

```python
import re
```

```python
emails=[]
for i in a:
    if(re.search('[a-z]+@[a-z]+.[a-z]+',i)):
        emails.append(re.search('[a-z]+@[a-z]+.[a-z]+',i).group())
    elif(re.search('[a-z]+\[at\][a-z]+.[a-z]+',i)):
        emails.append(re.search('[a-z]+\[at\][a-z]+.[a-z]+',i).group())
    
```

```python
emails
```

```python
email_body=pd.DataFrame(emails,columns=["email_in_body"])
```

```python
email_body.to_csv("Emails_in _body.csv")
```

# Task-3

Check and fetch mail number from the email body section if any present and store it in another CSV.



```python
phone=[]
for i in a:
    if(re.search('[0-9]+\-[0-9]+\-[0-9]+',i)):
        phone.append(re.search('[0-9]+\-[0-9]+\-[0-9]+',i).group())
    elif(re.search('\+[0-9]\s\([0-9]+\)\s[0-9]+\-[0-9]+',i)):
        phone.append(re.search('\+[0-9]\s\([0-9]+\)\s[0-9]+\-[0-9]+',i).group())
    elif(re.search('\([0-9]+\)\s[0-9]+\-[0-9]+',i)):
        phone.append(re.search('\([0-9]+\)\s[0-9]+\-[0-9]+',i).group())
        
```

```python
phone
```

```python
phone_number=pd.DataFrame(phone,columns=["phone_num_in_body"])
```

```python
phone_number
```

```python
phone_number.to_csv("phone_number_in _body.csv")
```
